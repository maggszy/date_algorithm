```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

First group

```python
x = np.linspace(0, 1, 100)
y1= np.power(x,0.999999)* np.log(x)
y2= np.power(10,7) * x
y3= np.power(1.000001,x) 
y4= np.square(x)
```

```python
plt.plot(x,y1, label="y1")
#plt.plot(x,y2, label="y2")
plt.plot(x,y3, label="y3")
plt.plot(x,y4, label="y4")
plt.grid()
plt.legend()
plt.show()
```
